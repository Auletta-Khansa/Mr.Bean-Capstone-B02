```python
# Definisikan path ke direktori dengan gambar
train_path = '/content/drive/MyDrive/Capstone-v17/Coffee-Bean-v17/'

# Definisikan path ke subdirektori yang berisi gambar
subdirectories = os.listdir(train_path)

# Ukuran target yang diinginkan
target_size = (224, 224)

# Iterasi melalui subdirektori
for subdirectory in subdirectories:
    subdirectory_path = os.path.join(train_path, subdirectory)
    image_files = os.listdir(subdirectory_path)

    # Iterasi melalui gambar dalam setiap subdirektori
    for image_file in image_files:
        image_path = os.path.join(subdirectory_path, image_file)
        image = cv2.imread(image_path)

        # Resize gambar ke ukuran target
        image = cv2.resize(image, target_size)

        # Simpan gambar yang sudah diresize
        cv2.imwrite(image_path, np.uint8(image))

print("Proses pengubahan ukuran dan normalisasi gambar selesai.")
```

### **Model MobileNetV2**

```python
# Import model
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2
```

```python
# Inisialisasi tinggi dan lebar gambar
IMG_HEIGHT = 224
IMG_WIDTH = 224

# Inisialisasi dimensi gambar
IMG_SHAPE = (IMG_HEIGHT, IMG_WIDTH, 3)

# Inisialisasi base model menggunakan MobileNetV2
base_model = MobileNetV2(input_shape=IMG_SHAPE, input_tensor=None,
                                                include_top=False,
                                                weights='imagenet')
base_model.trainable = True
```

```python
# Mendefinisikan model
model = tf.keras.Sequential()
model.add(base_model)
model.add(tf.keras.layers.GlobalAveragePooling2D())
model.add(tf.keras.layers.Dense(320, activation='relu'))
model.add(tf.keras.layers.Dropout(0.3))
model.add(tf.keras.layers.Dense(320, activation='relu'))
model.add(tf.keras.layers.Dropout(0.3))
model.add(tf.keras.layers.Dense(320, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)))
model.add(tf.keras.layers.Dropout(0.5))
model.add(tf.keras.layers.Dense(4, activation='softmax'))

# Kompilasi model
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.001, momentum=0.9),
      loss = tf.keras.losses.CategoricalCrossentropy(from_logits = True),
      metrics=['accuracy'])

# Ringkasan model
print(model.summary())
```

### **Model MobileNetV3Large**

```python
# Membuat base model menggunakan MobileNetV3Large
base_model = tf.keras.applications.MobileNetV3Large(include_top= False,)

# Membekukan base model agar tidak dapat dilatih lebih lanjut
base_model.trainable = False

# Membuat input ke dalam model
inputs = tf.keras.layers.Input(shape =(224, 224, 3), name = "input-layer")

# Rescaling
#x = tf.keras.layers.experimental.preprocessing.Rescaling(1/255.)(inputs)

# Memasukkan data input ke dalam base model
x = base_model(inputs)
print(f"Dimensi setelah input melewati base model: {x.shape}")

# Melakukan average pooling pada output base model
x = tf.keras.layers.GlobalAveragePooling2D(name = "global_average_pooling_layer")(x)
print(f"Shape after GlobalAveragePooling2D: {x.shape}")

# Menambahkan lapisan klasifikasi di atasnya
x = tf.keras.layers.Dense(320, activation='relu')(x)
x = tf.keras.layers.Dropout(0.3)(x)
x = tf.keras.layers.Dense(320, activation='relu')(x)
x = tf.keras.layers.Dropout(0.3)(x)
x = tf.keras.layers.Dense(320, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01))(x)
x = tf.keras.layers.Dropout(0.5)(x)

# Membuat lapisan aktivasi output
outputs = tf.keras.layers.Dense(4, activation = "softmax", name = "output-layer")(x)

# Menggabungkan input dan output ke dalam model utama
model_0 = tf.keras.Model(inputs, outputs)

# Kompilasi model utama
model_0.compile(loss = "categorical_crossentropy",
                optimizer = tf.keras.optimizers.SGD(learning_rate=0.001, momentum=0.9),
                metrics = ["accuracy","AUC"])

# Ringkasan model
print(model_0.summary())
```

### **Training Process**

```python
# Jumlah iterasi pelatihan model
epochs = 100

# Penyesuaian Learning Rate
# learning_rate_scheduler = tf.keras.callbacks.LearningRateScheduler(
#   lambda epoch: 1e-5 * 10**(epoch / 20))

# Early Stopping
# early_stopping = tf.keras.callbacks.EarlyStopping(
#    monitor='val_loss',
#    patience=10,
#    restore_best_weights=True)

# Menggunakan ReduceLROnPlateau untuk mengurangi learning rate ketika terjadi perubahan <min_delta> pada <val_loss> lebih dari <patience> epoch
reduce_learning_rate = tf.keras.callbacks.ReduceLROnPlateau(
                            monitor = 'val_loss',
                            mode = 'max',
                            min_delta = 0.01,
                            patience = 10,
                            factor = 0.25,
                            verbose = 1,
                            cooldown = 0,
                            min_learning_rate = 0.00000001)

# Checkpoint model terbaik selama pelatihan
checkpointer = ModelCheckpoint(
                  monitor='val_loss',
                  filepath='/content/drive/MyDrive/Capstone-v16/model_mobileNet_100_rlr.h5',
                  verbose=1,
                  save_best_only=True)

# Melatih model menggunakan data training dan validation
trained = model_0.fit(
            train_data,
            epochs=epochs,
            validation_data = test_data,
            callbacks=[checkpointer, reduce_learning_rate]) # Menggunakan callback untuk mengurangi learning rate
```

```python
# Membuat plot grafik accuracy model
plt.plot(trained.history['accuracy'])
plt.plot(trained.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

# Membuat plot grafik loss model
plt.plot(trained.history['loss'])
plt.plot(trained.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

# Membuat plot grafik AUC model
plt.plot(trained.history['auc'])
plt.plot(trained.history['val_auc'])
plt.title('Model auc')
plt.ylabel('Auc')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```
